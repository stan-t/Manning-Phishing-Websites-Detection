```python
# Import pandas and numpy
import pandas as pd
import numpy as np

# Fix the random seed
np.random.seed(7)
```

```python
# Load the dataset
data = pd.read_csv("Phishing.csv")
data.head(10).T
```

Now, let's quickly find out how the different classes in the dataset are distributed. We will use the `Counter` class for that. 

```python
# Import the Counter class
from collections import Counter
```

```python
classes = Counter(data["Result"].values)
classes.most_common()
```

Let's turn the above result into a nice DataFrame. 

```python
class_dist = pd.DataFrame(classes.most_common(), columns=['Class', 'Num_Observations'])
class_dist
```

We can prepare a plot for conveying the above information as well. That process is a bit more involved. 

```python
import matplotlib.pyplot as plt

# Make your plots appear within the notebook
%matplotlib inline

# Stylize the plots
plt.style.use("ggplot")
```

```python
subplot = class_dist.groupby("Class")["Num_Observations"].sum().plot(kind="barh", width=0.2, figsize=(10,8))

subplot.set_title("Class distribution of the websites", fontsize = 15)
subplot.set_xlabel("Number of Observations", fontsize = 14)
subplot.set_ylabel("Class", fontsize = 14)

for i in subplot.patches:
    subplot.text(i.get_width()+0.1, i.get_y()+0.1, \
            str(i.get_width()), fontsize=11)
```
